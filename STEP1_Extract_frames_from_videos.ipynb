# Step 3: Extract Frames from Video File

Import the necessary libraries

```python
import cv2
import os
```

Replace the name of the video file in the line of code below if you want to extract from another video file

```python
vidcap = cv2.VideoCapture('file.mp4')
```

Run the following code to extract the frames from the video file

```python
success,image = vidcap.read()
count = 0

folder = "/content/frames"

while success:
    filename = "frame%d.jpg" % count
    cv2.imwrite(os.path.join(folder , filename), image)     # save frame as JPEG file      
    success,image = vidcap.read()
  
    print('Read a new frame: ', success)
    count += 1
```
